```python
import pandas as pd
import numpy as np
```

# Open CSV

```python
dfRating=pd.read_csv('rating.csv')
dfRating=dfRating
dfRating.head()
```

# Pivot Table

```python
dfPivot=dfRating.pivot_table(
    index=['user_id'],columns=['anime_id'],values='rating'
)
dfPivot=dfPivot.fillna(0)
dfPivot.head()
```

# CorrMatrix

```python
corrMatrix=dfPivot.corr()
corrMatrix.head()
```

# Save Corr Matrix to Pickle

```python
import pickle
with open('corrMatrix.pkl','rb') as x:
        corrMatrix=pickle.load(x)
```

```python
corrMatrix=corrMatrix.set_index('anime_id')
corrMatrix.head()
```

```python
userX=[(91,5),(1,5),(48,4),(1,3)]
finalSkor=pd.DataFrame()
for anime_id,rating in userX:
    skor=corrMatrix.loc[anime_id]*rating
    skor=skor.sort_values(ascending=False)
    finalSkor=finalSkor.append(skor)
finalSkor=finalSkor.sum().sort_values(ascending=False)
finalSkor[:15].index.values
final=finalSkor[:8].index.values
final
```

```python
data=pd.read_csv('AnimeUpdatedImage.csv')
data.head()
```
